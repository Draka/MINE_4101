```python
# import gdal, install gdal $conda install gdal
from osgeo import gdal
import numpy as np 
import matplotlib.pyplot as plt
```

```python
import glob
rasternames = glob.glob("raster/*.tif")
```

```python
NoVar = 0
variable = rasternames[NoVar]

v1 = gdal.Open(variable)
col   = v1.RasterXSize
rows  = v1.RasterYSize

#Ejectutar solo para comprobar
v1val = v1.GetRasterBand(1).ReadAsArray().flatten()
NanValues = np.where(v1val == -9999)[0]
#v1val[NanValues]=None
fig, ax1 = plt.subplots(1,1)
cm1 = plt.cm.get_cmap('Accent') 
#im = plt.imshow(v1val.reshape(rows,col),cmap='tab20c')
plt.imshow(v1val.reshape(rows,col),cmap='jet')

cb = plt.colorbar()
cb.set_label(variable, color='black')
ax1.set_title(variable, color="Black")
```
